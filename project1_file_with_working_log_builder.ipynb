```python
# COMBAINING 2 DATAFRANES -STARTS HERE 

import pandas as pd

# Load the Excel file into a Pandas dataframe
df = pd.read_excel('geological_logs.xlsx', sheet_name='geo_logs')

# Make the values in the 'description' column lowercase
df['description'] = df['description'].str.lower()

# Make the values in the 'description2' column lowercase
df['description2'] = df['description2'].str.lower()

# Save the updated dataframe to a new Excel file
#df.to_excel('geological_logs2.xlsx', sheet_name='geo_logs_lower')



# Load the Excel file into a Pandas dataframe
df_rock_type = pd.read_excel('rock_types.xlsx', sheet_name='Sheet')


# create an empty column with default value as None
df["main rock type"] = None


for index, row in df.iterrows():
    # Check if the value in the 'description' column is a string
    if isinstance(row['description'], str):
        # Check if any of the rock types from df_rock_type are present in the description
        for rock_type in df_rock_type['Rock Type']:
            if rock_type in row['description']:
                # Write the matching rock type into the 'main rock type' column of df
                df.loc[index, 'main rock type'] = rock_type
                # Break out of the inner loop, since we have found a match
                break

                

# Save the updated dataframe to a new Excel file
df.to_excel('geological_logs5.xlsx', sheet_name='geo_logs_lower1')

# empty_rows = df[df['main rock type'].isna()].head(500)
# print(empty_rows)
empty_rows = df[df['main rock type'].isnull()].head(500)
print(empty_rows)
empty_rows.to_excel('empty_rows2.xlsx', index=False)
```

```python
import pandas as pd

def read_xlsx_file(file_path,sheet_name_input):
    # Get the filename without the extension
    file_name = file_path.split('.')[0]

    # Load the Excel file into a Pandas dataframe
    df = pd.read_excel(file_path,sheet_name_input )

    # Rename the columns to include the file name as a prefix
    columns = list(df.columns)
    new_columns = [f"{file_name}_{col}" for col in columns]
    df.columns = new_columns

    return df
```
